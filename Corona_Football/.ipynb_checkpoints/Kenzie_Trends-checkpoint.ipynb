```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
# multiTimeline.csv

# rading in the csv file

df_timeline = pd.read_csv('multiTimeline.csv')
df_timeline
```

```python
# displaying the columns

df_timeline.columns
```

```python
# renaming the columns 

df_timeline.columns = ['Week', 'Covid', 'corona', 'coronavirus', 'Football']
df_timeline.columns
```

```python
# dtatype for each column

df_timeline.dtypes
```

```python
# replacing '<1' with '1'

df_timeline['Covid'] = df_timeline['Covid'].str.replace('<','')
df_timeline['corona'] = df_timeline['corona'].str.replace('<','')
df_timeline['coronavirus'] = df_timeline['coronavirus'].str.replace('<','')
```

```python
# converting the object datatypes to string
df_timeline['Covid'] = df_timeline['Covid'].astype(str).convert_dtypes()
df_timeline['corona'] = df_timeline['corona'].astype(str).convert_dtypes()
df_timeline['coronavirus'] = df_timeline['coronavirus'].astype(str).convert_dtypes()
```

```python
# converting the strings datatypes to int
df_timeline['Covid'] = df_timeline['Covid'].astype('Int64')
df_timeline['corona'] = df_timeline['corona'].astype('Int64')
df_timeline['coronavirus'] = df_timeline['coronavirus'].astype('Int64')
```

```python
# changing the week to datetime

df_timeline['Week'] = pd.to_datetime(df_timeline['Week'])
```

```python
df_timeline.set_index('Week', inplace=True)
```

```python
df_timeline.plot(figsize=(20, 10))

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot()
plt.tick_params(rotation=45)
ax.xaxis.set_major_formatter(Week.DateFormatter('%y:%m:%d'))
plt.show()
```
